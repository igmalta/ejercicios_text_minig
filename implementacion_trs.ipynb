## Procesamiento del texto

```python
source = 'data/med/MED.ALL'
target = "data/med/med.dat"

#Chequear si el archivo del target ya existe
if os.path.exists(target):
    os.remove(target)
else:
    print("Target does not exist")

f = open(source, 'r') #SOURCE
f2= open(target,"w+") #TARGET

initialLine = 1 #COUNTER
lineNumber = initialLine # LINE NUMBER
for line in f:
    if line.startswith('.I'): #CHECK THAT WE ARE HAVE TO ADD A JUMP OF LINE 
        print('PARAGRAPH NUMBER:: ',lineNumber)
        if lineNumber != initialLine:
            f2.write("\n")#ADD A LINE JUMP
    elif not line.startswith('.W'): #CHECK TO KNOW IF WE ARE AT A CONTENT ROW
        print('CONTENT: ',lineNumber)
        tempLine =  line.rstrip('\n') #REMOVES JUMP OF LINE
        cleanLine = tempLine.strip() # REMOVES EMPTY SPACES
        addingSpace = cleanLine + " "  #ADDS A EMPTY SPACE AT THE LEFT
        f2.write(addingSpace)
    else: #NEXT LINE WE SHOULD FIND A PARAGRAPH
        print("PARAGRAPH IS GOING TO START", lineNumber)
    lineNumber += 1
        
f.close()
f2.close()
```

## Procesamiento de querys

```python
source = 'data/med/MED.QRY'
target = "data/med/med-queries.txt"

#Chequear si el archivo del target ya existe
if os.path.exists(target):
    os.remove(target)
else:
    print("Target does not exist")

f = open(source, 'r') #SOURCE
f2= open(target,"w+") #TARGET

initialLine = 1 #COUNTER
lineNumber = initialLine # LINE NUMBER
for line in f:
    if line.startswith('.I'): #CHECK THAT WE ARE HAVE TO ADD A JUMP OF LINE 
        print('PARAGRAPH NUMBER:: ',lineNumber)
        if lineNumber != initialLine:
            f2.write("\n")#ADD A LINE JUMP
    elif not line.startswith('.W'): #CHECK TO KNOW IF WE ARE AT A CONTENT ROW
        print('CONTENT: ',lineNumber)
        tempLine =  line.rstrip('\n') #REMOVES JUMP OF LINE
        cleanLine = tempLine.strip() # REMOVES EMPTY SPACES
        addingSpace = cleanLine + " "  #ADDS A EMPTY SPACE AT THE LEFT
        f2.write(addingSpace)
    else: #NEXT LINE WE SHOULD FIND A PARAGRAPH
        print("PARAGRAPH IS GOING TO START", lineNumber)
    lineNumber += 1
        
f.close()
f2.close()
```

## Indexación de los datos mediante formato de índice invertido

```python
inv_idx = metapy.index.make_inverted_index('med.toml')
```

### Resumen de medidas

```python
# Número de documentos en el corpus
inv_idx.num_docs()
```

```python
# Número de palabras únicas
inv_idx.unique_terms()
```

```python
# Largo promedio de cada documento
inv_idx.avg_doc_length()
```

```python
# Cantidad total de palabras del corpus
inv_idx.total_corpus_terms()
```

## Indexación de los datos mediante formato de índice forward

```python
fwd_idx = metapy.index.make_forward_index('med.toml')
```

## Ranking de documentos

```python
# Ranker elegido
ranker = metapy.index.DirichletPrior()

# Query de inetrés
query = metapy.index.Document()
query.content("chemotherapy of breast cancer")

# Top 5 documentos rankeados
top_docs = ranker.score(inv_idx, query, num_results=5)
top_docs
```

## Previsualización de los documentos mejor rankeados

```python
for num, (d_id, _) in enumerate(top_docs):
    content = inv_idx.metadata(d_id).get('content')
    print("{}. {}...\n".format(num + 1, content[0:250]))
```
